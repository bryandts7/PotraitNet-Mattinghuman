```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def detect_and_preprocess(image_path, output_size=(224, 224)):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    if len(faces) == 0:
        print("No face detected.")
        return None
    
    # Assume the first face is the one we want (in case of multiple detections)
    x, y, w, h = faces[0]
    
    # Crop the face region
    face_crop = img[y:y+h, x:x+w]
    
    # Resize the cropped face
    resized_face = cv2.resize(face_crop, output_size)
    
    resized_face_rgb = cv2.cvtColor(resized_face, cv2.COLOR_BGR2RGB)
    
    return resized_face_rgb

image_path = 'custom_image_1.jpg'  
preprocessed_image = detect_and_preprocess(image_path)

if preprocessed_image is not None:
    plt.imshow(preprocessed_image)
    plt.title("Preprocessed Face")
    plt.axis('off')
    plt.show()
```
